```python
from pandas import read_csv
import seaborn as sns
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2,f_regression
df=sns.load_dataset('tips')
df
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['smoker']=lb.fit_transform(df['smoker'])
df['sex']=lb.fit_transform(df['sex'])
df['day']=lb.fit_transform(df['day'])
df['time']=lb.fit_transform(df['time'])
```

```python
x = df.drop('tip',axis=1)
y = df.tip
x
```

```python
test = SelectKBest(score_func=f_regression, k=3)
fit = test.fit(x, y)
```

```python
test = SelectKBest(score_func=f_regression,k=3).fit(x,y)
test
```

```python
np.round(test.pvalues_,2)
```

```python
np.round(test.pvalues_,3)
```

```python
df.columns[test.get_support(indices=True)]
```

```python
df.columns[np.where(test.pvalues_>0.05)]
```
