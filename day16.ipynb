```python
from scipy.stats import zscore
import seaborn as sns
import numpy as np
data=sns.load_dataset('iris')
df=data.copy()

z_scores = np.abs(zscore(df.drop('species',axis=1)))
```

```python
z_scores
```

```python
z_scores.loc[10:15]
```

```python
threshold = 3
non_outliers = (z_scores < threshold).all(axis=1)
df_no_outliers = df[non_outliers]
```

```python
df_no_outliers
```

```python
threshold = 3
outliers = (z_scores > threshold)
outlier_rows = df[outliers.any(axis=1)]
outlier_rows
```

```python
from sklearn.ensemble import IsolationForest
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data =sns.load_dataset('iris')
data
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
data['species']=lb.fit_transform(data['species'])
```

```python
clf = IsolationForest(random_state=0,contamination=.01)
clf.fit(data)
```

```python
y_pred_outliers = clf.predict(data)
idx=np.where(y_pred_outliers==-1)
idx[0]
```

```python
data.drop(index=idx[0],axis=0,inplace=True)
```
